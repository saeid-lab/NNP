```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Fc(Rij,Rc):
    if Rij<= Rc:
        value = 0.5 * (np.cos(np.pi*(Rij/Rc)) +1)
    if Rij> Rc:
        value = 0
    
    return value
```

```python
def G4(zeta,lamda,theta,eta,Rij,Rik,Rjk):
    t1= (1+lamda * np.cos(np.deg2rad(theta)))**zeta
    t2=np.exp(-eta*(((Rij)**2) +((Rik)**2) + ((Rjk)**2)))
    t3=t1*t2
    return t3
```

```python
def G4_new(zeta,lamda,theta,eta,Rij,Rik,Rjk):
    t1= (1+lamda * np.cos(np.deg2rad(theta)))**zeta
    t2=np.exp(-eta*(((Rij)**2) +((Rik)**2) + ((Rjk)**2)))
    t3=t1*t2
    t4=Fc(Rij,10)*Fc(Rjk,10)*Fc(Rik,10)
    return t4*t3
```

```python
def G5(zeta,lamda,theta,eta,Rij,Rik):
    t1= (1+lamda * np.cos(np.deg2rad(theta)))**zeta
    t2=np.exp(-eta*(((Rij)**2) +((Rik)**2)))
    t3=t1*t2
    t4=Fc(Rij,12)*Fc(Rik,12)
    return t4*t3
```

```python
teta = range(0,361,1)
G4_1=[]
for i in teta:
    G4_1.append(G4(1,1,i,1,0,0,0) * (2**(1-1)) * Fc(0,10) * Fc(0,10)* Fc(0,10))
    
G4_2=[]
for i in teta:
    G4_2.append(G4(2,1,i,1,0,0,0) * (2**(1-2)) * Fc(0,10) * Fc(0,10)* Fc(0,10))
    
G4_4=[]
for i in teta:
    G4_4.append(G4(4,1,i,1,0,0,0) * (2**(1-4)) * Fc(0,10) * Fc(0,10)* Fc(0,10))
    
G4_16=[]
for i in teta:
    G4_16.append(G4(16,1,i,1,0,0,0) * (2**(1-16)) * Fc(0,10) * Fc(0,10)* Fc(0,10))
    
G4_64=[]
for i in teta:
    G4_64.append(G4(64,1,i,1,0,0,0) * (2**(1-64)) * Fc(0,10) * Fc(0,10)* Fc(0,10))

plt.figure(figsize=(7,5))
plt.plot(teta,G4_1, 'r', label='\u03B6=1')
plt.plot(teta,G4_2, 'g', label='\u03B6=2')
plt.plot(teta,G4_4, 'k', label='\u03B6=4')
plt.plot(teta,G4_16, 'b', label='\u03B6=16')
plt.plot(teta,G4_64, 'c', label='\u03B6=64')

plt.legend(loc='best')
plt.xlabel('\u03B8')
plt.ylabel('G4, \u03BB=1')
plt.margins(0)
plt.xticks(range(0,361,60))
plt.show()
```

```python
teta = range(0,361,1)
G4_new_1=[]
for i in teta:
    G4_new_1.append(G4_new(1,1,i,1,0,0,0) * (2**(1-1)))
    
G4_new_2=[]
for i in teta:
    G4_new_2.append(G4_new(2,1,i,1,0,0,0) * (2**(1-2)))
    
G4_new_4=[]
for i in teta:
    G4_new_4.append(G4_new(4,1,i,1,0,0,0) * (2**(1-4)))
    
G4_new_16=[]
for i in teta:
    G4_new_16.append(G4_new(16,1,i,1,0,0,0)  * (2**(1-16)))
    
G4_new_64=[]
for i in teta:
    G4_new_64.append(G4_new(64,1,i,1,0,0,0) * (2**(1-64)))

plt.figure(figsize=(7,5))
plt.plot(teta,G4_new_1, 'r', label='\u03B6=1')
plt.plot(teta,G4_new_2, 'g', label='\u03B6=2')
plt.plot(teta,G4_new_4, 'k', label='\u03B6=4')
plt.plot(teta,G4_new_16, 'b', label='\u03B6=16')
plt.plot(teta,G4_new_64, 'c', label='\u03B6=64')

plt.legend(loc='best')
plt.xlabel('\u03B8')
plt.ylabel('G4, \u03BB=1')
plt.margins(0)
plt.xticks(range(0,361,60))
plt.show()
```
