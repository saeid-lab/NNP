```python
import numpy as np
import matplotlib.pyplot as plt
# from math import *
```

```python
def Fc(Rij,Rc):
    if Rij<= Rc:
        value = 0.5 * (np.cos(np.pi*(Rij/Rc)) +1)
    if Rij> Rc:
        value = 0
    
    return value
```

```python
# G1 Plot

# Generate 100 number betwwen 0 and 14
Rij_num = np.linspace(0,14,num=1000)

# 5 List created
num11,num9,num7,num5,num3, = ([] for i in range(5))

# Rc Valuse
Rc_values = [11.3,9,7,5,3]

for numbers in Rij_num:
    num11.append(Fc(numbers,Rc_values[0]))
    num9.append(Fc(numbers,Rc_values[1]))
    num7.append(Fc(numbers,Rc_values[2]))
    num5.append(Fc(numbers,Rc_values[3]))
    num3.append(Fc(numbers,Rc_values[4]))
    

# Ploting...
plt.figure(figsize=(7,7))
plt.plot(Rij_num, num11,'m',label='Rc=11,3')
plt.plot(Rij_num, num9, 'r',label='Rc=9')
plt.plot(Rij_num, num7, 'g',label='Rc=7',)
plt.plot(Rij_num, num5, 'b',label='Rc=5',)
plt.plot(Rij_num, num3, 'c', label='Rc=3',)

plt.legend(loc='best')
plt.xlabel('Rij')
plt.ylabel('Fc')
plt.margins(0)
plt.show()
```

```python
data_11x = (np.array(Rij_num)[:-1] + np.array(Rij_num)[1:]) / 2

data_11y= np.diff(num11) / np.diff(Rij_num)
data_9y = np.diff(num9) / np.diff(Rij_num)
data_7y = np.diff(num7) / np.diff(Rij_num)
data_5y = np.diff(num5) / np.diff(Rij_num)
data_3y = np.diff(num3) / np.diff(Rij_num)

plt.figure(figsize=(7,7))
plt.plot(data_11x, data_11y,'m',label='Rc=11,3')
plt.plot(data_11x, data_9y,'r',label='Rc=9')
plt.plot(data_11x, data_7y,'g',label='Rc=7')
plt.plot(data_11x, data_5y,'b',label='Rc=5')
plt.plot(data_11x, data_3y,'c', label='Rc=3')

plt.legend(loc='best')
plt.xlabel('Rij')
plt.ylabel('d1 Fc')
# plt.margins(0)
plt.show()
```

```python
data_11xx = (np.array(data_11x)[:-1] + np.array(data_11x)[1:]) / 2

data_11yy = np.diff(data_11y) / np.diff(data_11x)
data_9yy = np.diff(data_9y) / np.diff(data_11x)
data_7yy = np.diff(data_7y) / np.diff(data_11x)
data_5yy = np.diff(data_5y) / np.diff(data_11x)
data_3yy = np.diff(data_3y) / np.diff(data_11x)

plt.figure(figsize=(7,7))
plt.plot(data_11xx,data_11yy,'m',label='Rc=11,3')
plt.plot(data_11xx,data_9yy,'r',label='Rc=9')
plt.plot(data_11xx,data_7yy,'g',label='Rc=7')
plt.plot(data_11xx,data_5yy,'b',label='Rc=5')
plt.plot(data_11xx,data_3yy,'c', label='Rc=3')

plt.legend(loc='best')
plt.xlabel('Rij')
plt.ylabel('d2 Fc')
# plt.margins(0)
plt.show()
```
